```python
import pydub
import os
from os import listdir
from os.path import isfile, join
import pandas as pd
```

```python
wav_filename_ls = []
wav_filesize_ls = []
trainscript_ls = []

fname = "are you married"
fname1 = "are_you_married"

rootDir = "/home/nam/Desktop/data_test/"
trainDir = "/home/nam/Desktop/data_test/test"
df_are_you_married = pd.DataFrame()
count = 1
path = join(rootDir, fname)
fileList = [f for f in listdir(path) if isfile(join(path, f))]
for i, fls in enumerate(fileList):
    filename = path+"/"+fls
    if os.path.isfile(filename):
        new_filename = trainDir+"/"+fname1+"_"+str(count)+".wav"
        
        sound = pydub.AudioSegment.from_wav(filename)
        new_sound = sound.set_frame_rate(16000).set_channels(1)
        new_sound.export(new_filename, format="wav")
        
        count = count + 1
        
        wav_filesize = os.path.getsize(new_filename)
        trainscript = fname
        
        wav_filename_ls.append(new_filename)
        wav_filesize_ls.append(wav_filesize)
        trainscript_ls.append(trainscript)

df_are_you_married["wav_filename"] = wav_filename_ls
df_are_you_married["wav_filesize"] = wav_filesize_ls
df_are_you_married["transcript"] = trainscript_ls
df_are_you_married["wav_filename"][1]
df_are_you_married.to_csv("are_you_married.csv", index = False)
```

```python
import soundfile

def convert(wavfile_name):
    data, samplerate = soundfile.read(wavfile_name)
    soundfile.write(wavfile_name, data, samplerate, subtype='PCM_16')
    return wavfile_name

train["wav_filename"] = train["wav_filename"].apply(convert)
dev["wav_filename"] = dev["wav_filename"].apply(convert)
test["wav_filename"] = test["wav_filename"].apply(convert)
```
